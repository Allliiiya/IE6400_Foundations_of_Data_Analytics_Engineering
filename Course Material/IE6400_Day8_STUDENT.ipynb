#### Exercise 5 Checking for Duplicates

```python
import pandas as pd

# Create a DataFrame with some duplicate rows
data = {
    'Name': ['John', 'Anna', 'John', 'Linda', 'Anna'],
    'Age': [28, 24, 28, 32, 24],
    'City': ['New York', 'Paris', 'New York', 'London', 'Paris']
}

df = pd.DataFrame(data)

# Check for duplicate rows in the DataFrame
print(df.duplicated())
```

```python
# Count the number of duplicate rows in the DataFrame
print(df.duplicated().sum())
```

```python
# Remove duplicate rows from the DataFrame
df = df.drop_duplicates()
df
```

```python
# Check for duplicate rows based on the 'Name' and 'Age' columns
print(df.duplicated(subset=['Name', 'Age']))

# Remove duplicate rows based on the 'Name' and 'Age' columns
df = df.drop_duplicates(subset=['Name', 'Age'])
```

#### Exercise 7 Performing EDA for Google Playstore Dataset

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('googleplaystore.csv')
df
```

```python
# Display the first few rows of the DataFrame
print(df.head())

# Information about DataFrame including the data types
print(df.info())

# Summary statistics of the DataFrame
print(df.describe(include='all'))
```

```python
# Count the number of missing values in each column
print(df.isna().sum())
```

```python
# Handle missing values (you can fill or drop based on the requirement)
df = df.dropna()
print(df.isna().sum())
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

```python
# Count of unique values for each categorical feature
print(df['Category'].value_counts())
print(df['Content Rating'].value_counts())

# Bar Plot for Category
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(5, 12))
sns.countplot(data=df, y='Category')
plt.show()
```

```python
# Histogram for numerical features like Rating, Reviews, and Price
df['Rating'].hist()
plt.show()

df['Reviews'].astype(int).hist()
plt.show()
```

```python
# Boxplot to understand the relationship between Category and Rating
fig, ax = plt.subplots(figsize=(5, 12))
sns.boxplot(data=df, x='Rating', y='Category')
plt.show()
```

```python
# Correlation matrix to understand the relationship between numerical variables
corr_matrix = df[['Rating', 'Reviews']].astype(float).corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

#### Exercise 8 Analyzing through Visualizations in Athlete Events Dataset

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('athlete_events.csv')
df
```

```python
# Display the first few rows of the DataFrame
print(df.head())

# Information about DataFrame including the data types
print(df.info())

# Summary statistics of the DataFrame
print(df.describe(include='all'))
```

```python
import matplotlib.pyplot as plt

# Histogram for Age, Height, and Weight
df['Age'].hist(bins=30)
plt.title('Age Distribution')
plt.show()

df['Height'].hist(bins=30)
plt.title('Height Distribution')
plt.show()

df['Weight'].hist(bins=30)
plt.title('Weight Distribution')
plt.show()
```

```python
import seaborn as sns

# Bar Plot for Sex, Team, and Medal
sns.countplot(data=df, x='Sex')
plt.title('Gender Distribution')
plt.show()

sns.countplot(data=df, x='Medal')
plt.title('Medal Distribution')
plt.show()
```

```python
# Boxplot to understand the relationship between Age, Height, Weight, and Medal
sns.boxplot(data=df, x='Medal', y='Age')
plt.title('Age Distribution by Medal')
plt.show()

sns.boxplot(data=df, x='Medal', y='Height')
plt.title('Height Distribution by Medal')
plt.show()

sns.boxplot(data=df, x='Medal', y='Weight')
plt.title('Weight Distribution by Medal')
plt.show()
```

```python
# Top 10 countries with the most medals
top_countries = df['Team'].value_counts().head(10)
top_countries.plot(kind='bar')
plt.title('Top 10 Countries by Number of Medals')
plt.show()
```

```python
# Number of Athletes participated over the years
athletes_over_years = df.groupby('Year')['ID'].nunique()
athletes_over_years.plot()
plt.title('Number of Athletes Over the Years')
plt.show()
```

```python
athletes_over_years = pd.DataFrame(athletes_over_years)
athletes_over_years.sort_values(by='Year', ascending=True)
```

#### Exercise 9 EDA on the Classifcation Dataset 

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df
```

```python
# Display the first few rows of the DataFrame
print(df.head())

# Information about DataFrame including the data types
print(df.info())

# Summary statistics of the DataFrame
print(df.describe())
```

```python
# Distribution of the target variable
sns.countplot(data=df, x='species')
plt.title('Distribution of Species')
plt.show()
```

```python
# Histogram for all numerical features
df.hist(figsize=(10, 8))
plt.show()
```

```python
# Pair plot for all features colored by species
sns.pairplot(df, hue='species')
plt.show()
```

```python
# Calculate the correlation matrix
corr_matrix = df.corr()

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Boxplot for each feature by species
features = iris.feature_names
for feature in features:
    sns.boxplot(data=df, x='species', y=feature)
    plt.title(f'Boxplot of {feature} by Species')
    plt.show()
```

#### Exercise 10 EDA on the Estimation Dataset 


    CRIM - per capita crime rate by town
    ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
    INDUS - proportion of non-retail business acres per town.
    CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
    NOX - nitric oxides concentration (parts per 10 million)
    RM - average number of rooms per dwelling
    AGE - proportion of owner-occupied units built prior to 1940
    DIS - weighted distances to five Boston employment centres
    RAD - index of accessibility to radial highways
    TAX - full-value property-tax rate per $10,000
    
    PTRATIO - pupil-teacher ratio by town
    B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
    LSTAT - % lower status of the population
    MEDV - Median value of owner-occupied homes in $1000's


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('BostonHousing.csv')
df
```

```python
# Display the first few rows of the DataFrame
print(df.head())

# Information about DataFrame including the data types
print(df.info())

# Summary statistics of the DataFrame
print(df.describe())
```

```python
# Histogram for the target variable (MEDV)
df['medv'].hist(bins=30)
plt.title('Distribution of MEDV')
plt.show()
```

```python
# Calculate the correlation matrix
corr_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize = (9,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plots for each feature vs. the target variable
for feature in df.columns:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=df, x=feature, y='medv')
    plt.title(f'Relationship between {feature} and MEDV')
    plt.show()
```

```python
# Pair plot for selected features
selected_features = ['rm', 'age', 'tax', 'lstat', 'medv']
sns.pairplot(df[selected_features])
plt.show()
```
