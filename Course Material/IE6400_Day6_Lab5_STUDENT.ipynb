#### Exercise 2.1 Iterating DataFrames with items( )

```python
import pandas as pd

dataframe = pd.DataFrame({
    'first_name' : ['John', 'Jane', 'Marry', 'Victoria', 'Gabriel', 'Layla'],
    'last_name' : ['Smith', 'Doe', 'Jackson', 'Smith', 'Brown', 'Martinez'],
    'age' : [34, 29, 37, 52, 26, 32]},
    index = ['id001', 'id002', 'id003', 'id004', 'id005', 'id006'])

print(dataframe)
```

```python
for col_name, data in dataframe.items():
    print("col_name: ", col_name, "\ndata: ", data)
```

```python
for col_name, data in dataframe.items():
    print("col_name: ", col_name, "\ndata: ", data[1])
```

```python
for col_name, data in dataframe.items():
    print("col_name: ", col_name, "\ndata: ", data['id002'])
```

axis=1 (Horizontal Concatenation):

    - When axis=1 is used in pd.concat(), it means that the concatenation will occur horizontally, along the columns.
    - It stacks the DataFrames side by side, aligning them by their indices. This is often used when you want to combine DataFrames with the same index but different columns.
    - The result is a larger DataFrame with columns from the original DataFrames concatenated together.

```python
import pandas as pd

df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame({'C': [7, 8, 9], 'D': [10, 11, 12]})

print(df1)
print(df2)
df_concatenated = pd.concat([df1, df2], axis=1)
df_concatenated
```

axis=0 (Vertical Concatenation):

    - When axis=0 is used in pd.concat(), it means that the concatenation will occur vertically, along the rows.
    - It stacks the DataFrames on top of each other, aligning them by their columns. This is useful when you want to append one DataFrame below another.

```python
import pandas as pd

df11 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df22 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

print(df11)
print(df22)
df_concatenated1 = pd.concat([df11, df22], axis=0)
df_concatenated1
```
