#### Excercise 1 Find the population variance and standard deviation

```python
# Importing necessary libraries that provide a comprehensive ecosystem 
# for scientific computing and data analysis in Python.
import matplotlib.pyplot as plt
from scipy.stats import t, norm
```

```python
# Data: Starting salaries for Corporation A
salaries = [41, 38, 39, 45, 47, 41, 44, 41, 37, 42]

# Calculate the number of data points
N = len(salaries)

print(salaries)
print(N)
```

```python
# Calculate the mean (average) salary
mean_val = sum(salaries) / N

# Calculate population variance from scratch
# - It measures how data points are dispersed around the mean.
# - It sums the squared differences between each salary and the mean, then divides by the number of data points.
p_variance = sum([(salary - mean_val) ** 2 for salary in salaries]) / N

# Calculate the population standard deviation from the calculated variance
# - Standard deviation is the square root of the variance.
std = sqrt(p_variance)
```

```python
# Print results calculated from scratch
print('# Formulas from scratch')
print('population variance =', round(p_variance, 3))
print('population standard deviation =', round(std, 3), end='\n\n')

# Calculate population variance and standard deviation using statistics module
# - The statistics module provides built-in functions for these calculations.
p_variance = pvariance(salaries)  # Calculate population variance using pvariance()
std = pstdev(salaries)  # Calculate population standard deviation using pstdev()

# Print results using statistics module
print('# Using pvariance, pstdev')
print('population variance with statistics =', round(p_variance, 3))
print('population standard deviation with statistics =', round(std, 3))
```

#### Exercise 3 Calculate Sample Means and Statistics

#### Exercise 5 Calculating and Visualizing a 95% Confidence Interval for a Population Mean ($\mu$)

```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Given values
sample_mean = 21.1
population_std_dev = 2
sample_size = 30
confidence_level = 0.95

# Z-score for 95% confidence
z_score = norm.ppf((1 + confidence_level) / 2)

# Calculate margin of error
margin_error = z_score * (population_std_dev / math.sqrt(sample_size))

# Calculate confidence interval
lower_limit = sample_mean - margin_error
upper_limit = sample_mean + margin_error

# Output the confidence interval
print(f"The 95% confidence interval for the population mean is ({lower_limit:.2f}, {upper_limit:.2f})")
```

```python
# Plotting the normal distribution
x = np.linspace(sample_mean - 4, sample_mean + 4, 100)
y = norm.pdf(x, sample_mean, population_std_dev / math.sqrt(sample_size))

plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.axvline(x=lower_limit, color='red', linestyle='--', label=f'Lower Limit ({lower_limit:.2f})')
plt.axvline(x=upper_limit, color='green', linestyle='--', label=f'Upper Limit ({upper_limit:.2f})')
plt.axvline(x=sample_mean, color='blue', label=f'Sample Mean ({sample_mean})')
plt.fill_between(x, y, 0, where=((x > lower_limit) & (x <= upper_limit)), color='gray', alpha=0.5)
plt.legend()
plt.title('95% Confidence Interval for Population Mean')
plt.xlabel('X')
plt.ylabel('Probability Density Function')
plt.show()
```

#### Exercise 6 Two-Sample Z-Test for Comparing Mean Credit Card Debts in California and Illinois

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Given values
x_bar1 = 3060  # Sample mean for California
x_bar2 = 2910  # Sample mean for Illinois
sigma1 = 960  # Standard deviation for California
sigma2 = 845  # Standard deviation for Illinois
n1 = 250  # Sample size for California
n2 = 250  # Sample size for Illinois

# Calculate the z-value
z_value = (x_bar1 - x_bar2) / np.sqrt((sigma1**2/n1) + (sigma2**2/n2))

# Calculate the critical z-values for alpha = 0.05 (two-tailed)
alpha = 0.05
z_critical_left = stats.norm.ppf(alpha/2)  # Left critical value
z_critical_right = stats.norm.ppf(1 - alpha/2)  # Right critical value

# Decision-making
if z_value < z_critical_left or z_value > z_critical_right:
    decision = "Reject the null hypothesis. There is a significant difference in the mean credit card debts between California and Illinois."
else:
    decision = "Fail to reject the null hypothesis. There isn't enough evidence to support a significant difference in the mean credit card debts between California and Illinois."

print(decision)
```

```python
# Visualization
x = np.linspace(-3.5, 3.5, 1000)
y = stats.norm.pdf(x, 0, 1)  # Standard normal distribution

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue')

# Shade the regions beyond the critical z-values
plt.fill_between(x, y, where=(x < z_critical_left), color='red', alpha=0.5)
plt.fill_between(x, y, where=(x > z_critical_right), color='red', alpha=0.5)

# Plot the computed z-value
plt.axvline(z_value, color='green', linestyle='--', label=f'z-value = {z_value:.2f}')
plt.axvline(z_critical_left, color='red', linestyle='--', label=f'Critical z-value (left) = {z_critical_left:.2f}')
plt.axvline(z_critical_right, color='red', linestyle='--', label=f'Critical z-value (right) = {z_critical_right:.2f}')

plt.title("Two-Sample Z-Test Visualization")
plt.xlabel("z-values")
plt.ylabel("PDF")
plt.legend()
plt.grid(True)
plt.show()
```

#### Exercise 10 Two-Sample T-Test for Comparing Mean Mathematics Test Scores of Students from Two Different Teachers

```python
# Importing necessary libraries
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Given values
smean1, sstd1, n1 = 473, 39.7, 8  # Teacher 1
smean2, sstd2, n2 = 459, 24.5, 18  # Teacher 2
alpha = 0.10  # Significance level

# Calculate the t-statistic and degrees of freedom
t = (smean1 - smean2) / math.sqrt((sstd1**2 / n1) + (sstd2**2 / n2))
df = ((sstd1**2 / n1 + sstd2**2 / n2)**2) / (((sstd1**2 / n1)**2 / (n1 - 1)) + ((sstd2**2 / n2)**2 / (n2 - 1)))

# Calculate the critical value from the t-distribution
t_critical = stats.t.ppf(1 - alpha/2, df)

# Decision
if abs(t) > t_critical:
    result = {
        "decision": "Reject the null hypothesis",
        "t": t,
        "t_critical": t_critical,
        "interpretation": "There is a significant difference in the mean mathematics test scores for the students of the two teachers."
    }
else:
    result = {
        "decision": "Do not reject the null hypothesis",
        "t": t,
        "t_critical": t_critical,
        "interpretation": "There is no significant difference in the mean mathematics test scores for the students of the two teachers."
    }

# Output the result
print(result)
```

```python
# Visualization
x = np.linspace(-4, 4, 100)
y = stats.t.pdf(x, df)
plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.axvline(x=t, color='red', linestyle='--', label=f'T = {t:.2f}')
plt.axvline(x=-t_critical, color='green', linestyle='--', label=f'T Critical = {-t_critical:.2f}')
plt.axvline(x=t_critical, color='green', linestyle='--')
plt.fill_between(x, y, 0, where=(x < -t_critical) | (x > t_critical), color='gray', alpha=0.5)
plt.legend()
plt.title('T-Test Visualization')
plt.xlabel('T')
plt.ylabel('Probability Density Function')
plt.show()
```

####  Exercise 12 Find the Confidence Interval for Population Mean Height Using T-Distribution

```python
# Importing necessary libraries
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Given values
x_bar, s, n, alpha = 168, 5, 15, 0.05  # Sample statistics and significance level

# Calculate the t critical value
t_critical = stats.t.ppf(1 - alpha/2, n-1)

# Calculate the margin of error
margin_error = t_critical * (s / math.sqrt(n))

# Calculate the confidence interval
lower_bound = x_bar - margin_error
upper_bound = x_bar + margin_error

# Output the confidence interval
print(f"The 95% confidence interval for the population mean height is ({lower_bound:.2f} cm, {upper_bound:.2f} cm)")
```

```python
# Visualization
x = np.linspace(x_bar - 3*margin_error, x_bar + 3*margin_error, 100)
y = stats.t.pdf(x, n-1, loc=x_bar, scale=s/math.sqrt(n))
plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.axvline(x=lower_bound, color='red', linestyle='--', label=f'Lower Bound = {lower_bound:.2f} cm')
plt.axvline(x=upper_bound, color='green', linestyle='--', label=f'Upper Bound = {upper_bound:.2f} cm')
plt.fill_between(x, y, 0, where=(x > lower_bound) & (x < upper_bound), color='gray', alpha=0.5)
plt.legend()
plt.title('95% Confidence Interval for Population Mean Height')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density Function')
plt.show()
```

#### Exercise 13 Two-Sample T-Test for Comparing Test Performances of Two Classes

#### Exercise 14 One-Sample Z-Test for Checking Machine Production Accuracy

```python
# Importing necessary libraries
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Given values
mu, sigma, n, x_bar, alpha = 50, 5, 30, 48.5, 0.05  # Population and sample statistics

# Calculate the z-statistic
z = (x_bar - mu) / (sigma / math.sqrt(n))

# Calculate the critical value from the z-distribution
z_critical = stats.norm.ppf(1 - alpha/2)

# Decision
if abs(z) > z_critical:
    result = {
        "decision": "Reject the null hypothesis",
        "z": z,
        "z_critical": z_critical,
        "interpretation": "There is evidence to suggest that the machine is not producing parts with the correct dimensions."
    }
else:
    result = {
        "decision": "Do not reject the null hypothesis",
        "z": z,
        "z_critical": z_critical,
        "interpretation": "There is no significant evidence to suggest that the machine is not producing parts with the correct dimensions."
    }

# Output the result
print(result)
```

#### Exercise 15 Paired T-Test for Assessing Training Program Effectiveness

```python
# Importing necessary libraries
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given values
before = [25, 30, 28, 35, 40]  # Productivity scores before the training
after = [30, 35, 32, 38, 42]  # Productivity scores after the training
alpha = 0.05  # Significance level

# Perform a paired t-test
t, p = stats.ttest_rel(after, before)

# Decision
if p < alpha:
    result = {
        "decision": "Reject the null hypothesis",
        "t": t,
        "p_value": p,
        "interpretation": "There is a significant difference in productivity levels before and after the training program."
    }
else:
    result = {
        "decision": "Do not reject the null hypothesis",
        "t": t,
        "p_value": p,
        "interpretation": "There is no significant difference in productivity levels before and after the training program."
    }

# Output the result
print(result)
```

#### Exercise 16 Chi-Square Test for Poker-Dealing Machine

```python
# Importing necessary libraries
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given values
observed = [404, 420, 400, 376]  # Observed frequencies
expected = [400, 400, 400, 400]  # Expected frequencies if suits are equally likely
alpha = 0.05  # Significance level

# Perform a chi-square test
chi2, p = stats.chisquare(observed, expected)

# Decision
if p < alpha:
    result = {
        "decision": "Reject the null hypothesis",
        "chi2": chi2,
        "p_value": p,
        "interpretation": "The suits are not equally likely."
    }
else:
    result = {
        "decision": "Do not reject the null hypothesis",
        "chi2": chi2,
        "p_value": p,
        "interpretation": "There is no significant evidence to suggest that the suits are not equally likely."
    }

# Output the result
print(result)
```

#### Exercise 17 F-Test for Comparing Variances

```python
# Importing necessary libraries
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

# Given values
group_a = [23, 25, 21, 22, 26, 28, 35, 36]  # Group A data
group_b = [42, 40, 41, 43, 45, 44, 46, 47]  # Group B data
alpha = 0.05  # Significance level

# Calculate variances
var_a = np.var(group_a, ddof=1)
var_b = np.var(group_b, ddof=1)

# Perform an F-test
F = var_a / var_b  # Calculate F statistic
df1 = len(group_a) - 1  # Degrees of freedom for group A
df2 = len(group_b) - 1  # Degrees of freedom for group B
p = 1 - stats.f.cdf(F, df1, df2)  # Calculate p-value

# Decision
if p < alpha:
    result = {
        "decision": "Reject the null hypothesis",
        "F": F,
        "p_value": p,
        "interpretation": "There is a significant difference in the variances of the two groups."
    }
else:
    result = {
        "decision": "Do not reject the null hypothesis",
        "F": F,
        "p_value": p,
        "interpretation": "There is no significant difference in the variances of the two groups."
    }

# Output the result
print(result)
```

```python
# Visualization
plt.figure(figsize=(10, 5))
plt.boxplot([group_a, group_b], labels=['Group A', 'Group B'])
plt.title('Boxplot of Group A and Group B')
plt.ylabel('Values')
plt.show()
```
