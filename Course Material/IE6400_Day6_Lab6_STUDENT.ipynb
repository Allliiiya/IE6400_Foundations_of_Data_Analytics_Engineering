#### Exercise 3.3 Filling Missing Values with Column Mean

```python
import pandas as pd

# Cleansing the missing values
missing_values = ["n.a.", "NA", "n/a", "na", 0]
df  = pd.read_csv('employee.csv', na_values = missing_values)

# Removing all NaN
df.dropna(axis = 0, inplace = True, how = 'all')

# Show df
df
```

```python
# Using mean
df['Salary'].fillna(int(df['Salary'].mean()), inplace = True)
```

```python
df
```

#### Exercise 3.9 Fill Missing DataFrame Values with Interpolation - polynomial
