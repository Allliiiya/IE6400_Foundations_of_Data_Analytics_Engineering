#### Exercise 6 Annotating a Bar Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

diamonds_df = sns.load_dataset('diamonds')
diamonds_df
```

```python
# Generate a bar plot using catplot function of the seaborn library:
ax = sns.catplot(x = "cut", 
                 data = diamonds_df, 
                 aspect = 1.5, 
                 kind = "count",
                 color = "blue")
```

```python
# get records in the DataFrame corresponding to ideal cut
ideal_group = diamonds_df.loc[diamonds_df['cut']=='Ideal']

# get the location of x coordinate where the annotation has to be placed
x = ideal_group.index.tolist()[0]

# get the location of y coordinate where the annotation has to be placed
y = len(ideal_group)

# print the x and y co-ordinates:
print(x)
print(y)
```

```python
# annotate the plot with any note or extra information

# Generate a bar plot using catplot function of the seaborn library:
ax = sns.catplot(x = "cut", 
                 data = diamonds_df, 
                 aspect = 1.5, 
                 kind = "count",
                 color = "blue")

plt.annotate('excellent polish and symmetry ratings;\nreflects almost all the light that enters it', 
             xy = (x,y), 
             xytext = (x + 0.3, y + 2000),
             arrowprops = dict(facecolor = 'red'))

plt.show()
```

#### Exercise 14 Histogram Revisited

```python
 # histogram using seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

mpg_df = sns.load_dataset("mpg")
ax = sns.histplot(mpg_df.weight, 
                  bins = 50, 
                  kde = False)

# `label` defines the name used in legend
plt.axvline(x = np.mean(mpg_df.weight), 
            color = 'red', 
            label = 'mean')
plt.axvline(x = np.median(mpg_df.weight), 
            color = 'orange',
            label = 'median')
plt.legend(loc = 'upper right')
plt.show()
```

#### Exercise 15 Creating and Exploring a Static Box Plot

#### Exercise 16 Creating a Static Violin Plot

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

mpg_df = sns.load_dataset("mpg")

# creating the feature 'model_decade'
mpg_df['model_decade'] = np.floor(mpg_df.model_year/10)*10
mpg_df['model_decade'] = mpg_df['model_decade'].astype(int)

# code for violinplots
sns.violinplot(x = 'model_decade', 
               y = 'mpg', 
               data = mpg_df, 
               hue = 'origin')
plt.show()
```

#### Exercise 17 Correlation Plots

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

file = pd.read_csv('AmesHousing.csv')
# Dropping some variables that aren't important
file.drop(['Order', 'PID'], axis = 1, inplace = True)

fig, ax = plt.subplots(figsize = (10, 14))
saleprice_corr = file.corr()[['SalePrice']]
sns.heatmap(saleprice_corr,
          annot = True,
          ax = ax)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

file = pd.read_csv('AmesHousing.csv')
# Dropping some variables that aren't important
file.drop(['Order', 'PID'], axis = 1, inplace = True)

fig, ax = plt.subplots(figsize = (10, 14))
saleprice_corr = file.corr()[['SalePrice']]

# Sorting 
sorted_saleprice_corr = saleprice_corr.sort_values(by = 'SalePrice', 
                                                   ascending = False)

sns.heatmap(sorted_saleprice_corr, annot = True, ax = ax)
plt.show()
```
